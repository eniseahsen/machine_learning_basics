```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error
```

```python
df = pd.read_csv(r"C:\db\insurance.csv")
```

```python
df.head()
```

```python
df = pd.get_dummies(df,columns=["sex","smoker","region"],drop_first=True)
```

```python
y =df[["charges"]]
x = df.drop("charges",axis=1)
```

```python
df.info
```

```python
lm = LinearRegression()
model=lm.fit(x,y)
```

```python
model.score(x,y)
```

```python
df
```

```python
model.predict([[19,26,0,1,1,0,0,1]])
```

```python
df_hata = pd.DataFrame()
```

```python
df_hata["y"]=y
df_hata
```

```python
y_tahmin = model.predict(x)
```

```python
df_hata["tahmin"]=y_tahmin
```

```python
df_hata
```

```python
df_hata["error"]=y-y_tahmin
```

```python
df_hata.head(10)
```

```python
df_hata["squared_error"]= df_hata["error"]**2
```

```python
df_hata
```

```python
df_hata["abs_error"] = np.abs(df_hata["error"])
```

```python
df_hata
```

```python
df_hata["percent_error"]=np.abs((y-y_tahmin)/y)
```

```python
df_hata
```

```python
df_hata.mean()
```

```python
mean_squared_error(y,y_tahmin)
```

```python
mean_absolute_error(y,y_tahmin)
```
